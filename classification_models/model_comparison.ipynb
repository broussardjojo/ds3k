```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid", palette="pastel")
```

```python
random_forest_best_score_train = 0.6756038647342996
cnb_best_score_train = 0.546060606060606
knn_best_score_train = 0.5739320920043812
```

```python
random_forest_test = 0.5942028985507246
cnb_test = 0.5120772946859904
knn_test = 0.5458937198067633
```

```python
records = [
    {'Model type': "Complement Naive Bayes", "score": cnb_best_score_train, "Score Type": 'train'},
	{'Model type': "Complement Naive Bayes", "score": cnb_test, "Score Type": 'test'},
	{'Model type': "Random Forest", "score": random_forest_best_score_train, "Score Type": 'train'},
	{'Model type': "Random Forest", "score": random_forest_test, "Score Type": 'test'},
	{'Model type': "KNN", "score": knn_best_score_train, "Score Type": 'train'},
	{'Model type': "KNN", "score": knn_test, "Score Type": 'test'},
]
records = pd.DataFrame.from_records(records)
```

```python
sns.barplot(x='Model type', y="score", data=records, hue='Score Type', )
plt.suptitle("Model type vs. Accuracy Score")
plt.savefig("model_comparison.png")
```
