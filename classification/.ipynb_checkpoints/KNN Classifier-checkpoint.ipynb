```python
#imports
import pandas as pd
import pathlib
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
%matplotlib inline
import utils.encoding
```

```python
#path
REPO_ROOT = pathlib.Path.cwd().parent
print(REPO_ROOT)
```

```python
#load training data that jojo created
library_train = pd.read_csv(REPO_ROOT / "data/spotify/group/mutually_exclusive_train_test/library_train.csv", index_col=0)
library_test = pd.read_csv(REPO_ROOT / "data/spotify/group/mutually_exclusive_train_test/library_test.csv", index_col=0)
```

```python
#drop unneeded columns
library_train = library_train.drop(columns=["artist", "track"])
library_test = library_test.drop(columns=["artist", "track"])
```

```python
library_train.head(10)
```

```python
library_train[library_train['label'] == 'nick']
```

```python
train_x, train_y = library_train.drop(columns='label'), library_train['label']
test_x, test_y = library_test.drop(columns='label'), library_test['label']
```

# Training and Hyperparameter Optimization <br>
K-Nearest Neighbors Classifier: [K-Nearest Neighbors](https://scikit-learn.org/stable/modules/neighbors.html)

```python
# hyperparemeters to be tuned
parameters_KNN = {
    'n_neighbors': (1,10, 1),
    'leaf_size': (20,40,1),
    'p': (1,2),
    'weights': ('uniform', 'distance'),
    'metric': ('minkowski', 'chebyshev')
}
```

```python
#build model and create grid search
knn = KNeighborsClassifier()
grid_search_KNN = GridSearchCV(estimator=knn, param_grid=parameters_KNN, scoring = 'accuracy', n_jobs = -1, cv = 5)
grid_search_KNN.fit(train_x, train_y)
```

```python
#looking at best parameters
print(grid_search_KNN.best_params_)
print(grid_search_KNN.best_score_)
```

```python
#create model with best params from grid search
optimizedKNN = KNeighborsClassifier(**grid_search_KNN.best_params_)
optimizedKNN.fit(train_x,train_y)
```

```python
#scoring
print(f"The KNN model's prediction of rings has a r^2 of {optimizedKNN.score(test_x,test_y): .2f}")
```

```python
## not sure if its an issue but richard has only 74 samples in the train_y set
train_y.value_counts()
```

```python
# classification report
y_pred = optimizedKNN.predict(test_x)
y_pred
target_names = ['jojo', 'nick','richard']
print(classification_report(test_y, y_pred, target_names = target_names))
```

```python
# plot conf matrix
conf_mat = confusion_matrix(y_true=test_y, y_pred=y_pred)

conf_mat_disp = ConfusionMatrixDisplay(conf_mat, display_labels=np.unique(y_pred))

# plot
conf_mat_disp.plot()
plt.title('K=10 NN Classifier Confusion Matrix')
plt.gcf().set_size_inches(10, 10)
plt.xticks(rotation = 90)
plt.grid(False)
```
